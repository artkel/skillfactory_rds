# DATA PARSING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from bs4 import BeautifulSoup
import requests
import re
import time

from datetime import datetime
```

```python
test = pd.read_csv('data/test.csv')
```

```python
test.shape
```

```python
test.brand.value_counts()
```

```python
brands_to_parse = test.brand.value_counts().index.tolist()
```

```python
# we will only parse data for the auto brands from our test dataset
brands_to_parse
```

```python
links = []
productionDate = []

now = datetime.now()
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
print("Start Date & Time =", dt_string)

for brand in brands_to_parse:
    url = f'https://auto.ru/cars/{brand}/used/'
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, 'html.parser')
    
    last_page_num = int(soup.find('span', class_="ControlGroup ControlGroup_responsive_no ControlGroup_size_s ListingPagination-module__pages").text[-2:])
    
    print(f'Brand: {brand}')
    print('***' * 14)
    time.sleep(.25)
    
    # for num in range(1, 4):
    for num in range(1, last_page_num+1):
        url = f'https://auto.ru/cars/{brand}/used/?page={num}'
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
        response.encoding = 'utf-8'
        soup = BeautifulSoup(response.text, 'html.parser')
        autolist = soup.find_all('div', class_='ListingItem-module__description')
        
        print(f'Page {num}')
        print('==' * 5)

        time.sleep(0.15)

        for item in autolist:
            productionDate.append(item.find('div', class_="ListingItem-module__year").text)
            for link in item.find_all('a', class_="Link ListingItemTitle-module__link", href=True):
                links.append(link['href'])


print()                
now = datetime.now()
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
print("Finish Date & Time =", dt_string)
print()
print(len(links))
print(len(productionDate))
```

```python
train_list = []
count = 0

now = datetime.now()
print("Start Date & Time =", now.strftime("%d/%m/%Y %H:%M:%S"))
print()

for link in links: 
    response = requests.get(link, headers={'User-Agent': 'Mozilla/5.0'})
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, 'html.parser')
    
    count += 1
    
    try:
        bodyType = soup.find('li', class_='CardInfoRow CardInfoRow_bodytype').find_all('span', class_="CardInfoRow__cell")[1].text    
    except:
        bodyType = np.NaN
    try:    
        brand = soup.find('h1').text.split(' ')[0].upper()
        model_name = soup.find('h1').text.split(' ')[1:]
    except:
        brand = np.NaN
        model_name = np.NaN      
    try:
        color = soup.find('li', class_='CardInfoRow CardInfoRow_color').find_all('span', class_="CardInfoRow__cell")[1].text
    except:
        color = np.NaN
    try:
        engineDisplacement = soup.find('li', class_='CardInfoRow CardInfoRow_engine').find_all('span', class_="CardInfoRow__cell")[1].text.split('/')[0].strip()
        enginePower = soup.find('li', class_='CardInfoRow CardInfoRow_engine').find_all('span', class_="CardInfoRow__cell")[1].text.split('/')[1].strip().replace('\xa0','')
        fuelType = soup.find('li', class_='CardInfoRow CardInfoRow_engine').find_all('span', class_="CardInfoRow__cell")[1].text.split('/')[2].strip()
    except:
        engineDisplacement = np.NaN
        enginePower = np.NaN
        fuelType = np.NaN
    try:
        mileage = soup.find('li', class_='CardInfoRow CardInfoRow_kmAge').find_all('span', class_="CardInfoRow__cell")[1].text.replace('\xa0','')
    except:
        mileage = np.NaN
    try:
        vehicleTransmission = soup.find('li', class_='CardInfoRow CardInfoRow_transmission').find_all('span', class_="CardInfoRow__cell")[1].text
    except:
        vehicleTransmission = np.NaN
    try:    
        customs = soup.find('li', class_='CardInfoRow CardInfoRow_customs').find_all('span', class_="CardInfoRow__cell")[1].text
    except:
        customs = np.NaN
    try:
        ownersCount = soup.find('li', class_='CardInfoRow CardInfoRow_ownersCount').find_all('span', class_="CardInfoRow__cell")[1].text
    except:
        ownersCount = np.NaN
    try:
        pts = soup.find('li', class_='CardInfoRow CardInfoRow_pts').find_all('span', class_="CardInfoRow__cell")[1].text
    except:
        pts = np.NaN
    try:
        drive = soup.find('li', class_='CardInfoRow CardInfoRow_drive').find_all('span', class_="CardInfoRow__cell")[1].text
    except:
        drive = np.NaN
    try:
        wheel = soup.find('li', class_='CardInfoRow CardInfoRow_wheel').find_all('span', class_="CardInfoRow__cell")[1].text
    except:
        wheel = np.NaN
    try:
        state = soup.find('li', class_='CardInfoRow CardInfoRow_state').find_all('span', class_="CardInfoRow__cell")[1].text
    except:
        state = np.NaN
    try:
        price = soup.find('span', class_="OfferPriceCaption PriceUsedOffer-module__caption").text.replace('\xa0','')
    except:
        price = np.NaN
    
    train = {
             'brand': brand,
             'model_name': model_name,
             'body_type': bodyType,
             'mileage': mileage,
             'color': color,
             'en_vol': engineDisplacement,
             'en_power': enginePower,
             'fuel': fuelType,
             'transmission': vehicleTransmission,
             'drive': drive,
             'prod_year': np.NaN,
             'num_doors': np.NaN,
             'vendor': np.NaN,    
             'num_owners': ownersCount,
             'pts': pts,
             'customs': customs,
             'wheel': wheel,
             'state': state,
             'price': price }
    
    train_list.append(train)

    if count % 1000 == 0:
        print()
        print('=============  -' + str(count) + '-  =============')
        now = datetime.now()
        print("Current Time =", now.strftime("%H:%M:%S"))
        print(f'Saving: {brand}')
        print()
        
    time.sleep(.08)


print()   
now = datetime.now()
print("Finish Date & Time =", now.strftime("%d/%m/%Y %H:%M:%S"))
```

## INITIAL CLEANING & SAVING DATA

```python
train_df = pd.DataFrame(train_list)
train_df['prod_year'] = productionDate
# train_df['vendor'] = vens
train_df.head()
```

```python
train_df.shape
```

```python
train_df.isnull().sum()
```

```python
# drop rows with missing values
cols_with_nans = ['model_name', 'body_type', 'mileage', 'en_vol', 'num_owners', 'state', 'customs', 'price']
train_df.dropna(axis=0, subset=cols_with_nans, inplace=True)
```

```python
# extrac & save model name
train_df['model_name'] = train_df.model_name.apply(lambda x: x[0])
```

```python
print(train_df.duplicated().sum())
```

```python
# drop duplicates
train_df.drop_duplicates(inplace=True)
```

```python
# save training data
train_df.to_csv(path_or_buf='data/train_V2.csv', index=False)
```
